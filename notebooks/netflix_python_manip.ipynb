```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the countries CSV file
countries = pd.read_csv('Countries.csv', sep=';', encoding='ISO-8859-1', on_bad_lines='skip')

# Combine all country columns into a single column
country_columns = ['country_1', 'country_2', 'country_3', 'country_4', 'country_5', 
                   'country_6', 'country_7', 'country_8', 'country_9', 'country_10', 
                   'country_11', 'country_12']

# Stack the country columns and remove NaN values
all_countries = countries[country_columns].stack().reset_index(drop=True)

# Count the occurrences of each country
country_counts = all_countries.value_counts()

# Select the top 10 countries with the most productions
top_countries = country_counts.head(10)

# Plot the bar chart with updated syntax
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.index, y=top_countries.values, hue=None, palette='viridis', legend=False)

# Add title and labels
plt.title('Top 10 Countries with Most Productions', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Number of Productions', fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

```python
# Count the distribution of production types (Movie vs TV Show)
production_types = netflix_titles['type'].value_counts()

# Plot a pie chart to show the distribution of Movie vs TV Show
plt.figure(figsize=(8,8))
production_types.plot(kind='pie', autopct='%1.1f%%', colors=["#f39c12", "#2ecc71"], startangle=90)
plt.title('Distribution of Movie and TV Show Productions')
plt.ylabel('')  # To remove the ylabel (it looks cleaner)
plt.show()

```

```python
# Count the distribution of ratings
rating_counts = netflix_titles['rating'].value_counts()

# Plot a bar chart for the distribution of ratings
plt.figure(figsize=(12,6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='Set3')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Productions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Clean the 'duration' column to get numeric values for movies' duration (TV Shows will be NaN)
netflix_titles['duration_numeric'] = pd.to_numeric(netflix_titles['duration'], errors='coerce')

# Plot a bar chart to show the distribution of movie durations
plt.figure(figsize=(12,6))
sns.histplot(netflix_titles['duration_numeric'].dropna(), bins=30, color='skyblue', kde=True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
